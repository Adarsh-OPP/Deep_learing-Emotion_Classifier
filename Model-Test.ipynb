```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt




model = tf.keras.models.load_model("emotion_model.h5")#change to model path


img = image.load_img("Tensorflow/Emotion_Classifier/TEST/WhatsAppImage2025-08-07at4,17,57PM (1)-photoaidcom-greyscale.png")#change to image path  |  There will be be test photo in test folder in /kaggle/input/human-emotions-datasethes/Emotions Dataset/Emotions Dataset/test
img_array = image.img_to_array(img)


img_array = np.expand_dims(img_array, axis=0)

# Predict
pred = model.predict(img_array)
pred_class = np.argmax(pred)
class_labels = ["Angry","Happy","Sad"]
print("Predicted emotion:", class_labels[pred_class])
print(pred)

plt.imshow(img)
plt.axis('off')
plt.show()
```
